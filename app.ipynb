```python
import gradio as gr
import pickle

# Load your pre-trained linear regression model
with open(r'A:/Python/OPS/model_io.pkl', 'rb') as f:
  model = pickle.load(f)

```

```python
def predict_marks(num_courses, time_spent):
  """Predicts marks based on number of courses and time spent."""
  prediction = model.predict([[num_courses, time_spent]])[0]

  if prediction >35:
      message = f"You Shall  Pass!: {prediction}"
  else:
    message = f"You Shall  not Pass! {prediction}"

  return message


iface = gr.Interface(
    fn=predict_marks,
    inputs=[
        gr.Number(label="Number of Courses"),
        gr.Number(label="Time Spent (hours)"),
    ],
    outputs="text",
    title="Marks Prediction",
    description="Predict your marks based on the number of courses and time spent studying.",
)

iface.launch()
```
